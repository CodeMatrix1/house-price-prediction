#Identifying null values

```python
null_percent=(df.isnull().sum() * 100 / len(df)).sort_values(ascending=False)
```

```python
null_percent[:20]
```

```python
#column drop due to missing values
threshold=40
cols_to_drop = null_percent[null_percent > threshold].index
df=df.drop(columns=cols_to_drop)
```

```python
df.shape
```

#Dropping Near Unique Cols

```python
categorical_cols=df.select_dtypes(include=['object']).columns
```

```python
near_unique_cols=[]
unique_threshold=0.95
n=len(df)
for col in categorical_cols:
    uniqueness_ratio = df[col].nunique(dropna=False) / n
    if uniqueness_ratio >= unique_threshold:
        near_unique_cols.append(col)
```

```python
df.columns
```

```python
near_unique_cols+=['Id']
```

```python
near_unique_cols
```

```python
df=df.drop(columns=near_unique_cols)
```

```python
df.shape
```

#Droping low info columns

```python
hard_threshold = 0.95
to_be_checked_threshold=0.85

low_info_cols = []
to_check_cols=[]
for col in df.columns:
    top_freq = df[col].value_counts(normalize=True, dropna=False).values[0]
    if top_freq > hard_threshold:
        low_info_cols.append(col)
    if to_be_checked_threshold<top_freq<hard_threshold:
        to_check_cols.append(col)
```

```python
low_info_cols
```

```python
to_check_cols
```

```python
df[to_check_cols].describe()
```

```python
categorical_cols=df.select_dtypes(include=['object']).columns
for feat in to_check_cols:
    if feat in categorical_cols:
        print(f"{feat}:\n",df[feat].value_counts(),df.groupby(feat)["SalePrice"].mean())
```

```python
df['SalePrice'].mean()
```

```python
categorical_cols=df.select_dtypes(include=['object']).columns
overall_mean=df['SalePrice'].mean()
spread_threshold=25000/overall_mean
potential_cols=[]
for feat in to_check_cols:
    if feat in categorical_cols:
        means=df.groupby(feat)["SalePrice"].mean()
        spread=(means.max()-means.min())/overall_mean
        if spread<=spread_threshold:
            potential_cols.append(feat)
```

```python
potential_cols
```

```python
low_info_cols+=potential_cols
```

```python
df = df.drop(columns=low_info_cols)
```

#Univariate Analysis

```python
# Separate features
num_cols = df.select_dtypes(include=['number']).columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in num_cols:
    if col == 'SalePrice':
        continue  # skip target if needed
    bins = min(50, max(10, int(np.sqrt(len(df[col])))))
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))  # Fixed size
    df[col].plot(kind='hist', bins=30, ax=axes[0], title=f'Histogram of {col}')
    df[col].plot(kind='kde', ax=axes[1], title=f'Density of {col}')
    plt.tight_layout()
    plt.show()
    print("skewness:",df[col].skew())
```

```python

for col in cat_cols:
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
